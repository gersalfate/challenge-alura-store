# 1. Análisis de facturación



```python
def ingreso(tienda):
  ingreso = tienda['Precio'].sum()
  return ingreso
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
datos = [tienda1, tienda2, tienda3, tienda4]
```

```python
ingresos = {}

for i in range(len(tiendas)):
  ingresos[tiendas[i]] = float(ingreso(datos[i]))

df_ingresos = pd.DataFrame.from_dict(ingresos, orient='index', columns=['Ingresos totales'])
df_ingresos.style.format('{:.2f}')
```

```python
import matplotlib.pyplot as plt

df_ingresos.plot(kind='bar', color='purple', legend=False, title='Ingresos totales (USD)', rot=0).ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.savefig('ingresos_totales.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 2. Ventas por categoría

```python
def producto_tienda(tienda):
  producto = tienda['Categoría del Producto'].value_counts().sort_values(ascending=False)
  return producto
```

```python
productos = []

for i in range(len(tiendas)):
  productos.append(producto_tienda(datos[i]))
  productos[i] = productos[i].to_frame(name=f'Total {tiendas[i]}')

productos = pd.concat(productos, axis=1)
productos
```

```python
productos.plot(kind='pie', subplots=True, layout=(2,2), figsize=(12,8), legend=False, ylabel='', title=tiendas, autopct='%1.1f%%')
plt.suptitle('Distribución de productos vendidos por categoría')
plt.tight_layout()
plt.show()
```

```python
def venta_tienda(tienda):
  venta = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
  return venta
```

```python
ventas = []

for i in range(len(tiendas)):
  ventas.append(venta_tienda(datos[i]))
  ventas[i] = ventas[i].to_frame(name=f'Total {tiendas[i]}')

ventas = pd.concat(ventas, axis=1)
ventas
```

```python
import matplotlib.pyplot as plt

ventas.plot(kind='pie', subplots=True, layout=(2,2), figsize=(12,8), legend=False, ylabel='',
            title=tiendas, autopct='%1.1f%%', textprops={'fontsize': 7})

plt.suptitle('Distribución de ventas por categoría')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

lista_procesada = []

for nombre, df in zip(tiendas, datos):
  temp_df = df.copy()
  temp_df['Tienda'] = nombre
  temp_df['Fecha de Compra'] = pd.to_datetime(temp_df['Fecha de Compra'], dayfirst=True)
  lista_procesada.append(temp_df)

df_total = pd.concat(lista_procesada)

df_mensual = df_total.groupby(['Tienda', df_total['Fecha de Compra'].dt.to_period('M')])['Precio'].sum().reset_index()
df_mensual['Fecha'] = df_mensual['Fecha de Compra'].dt.to_timestamp()

df_mensual['Media_6M'] = df_mensual.groupby('Tienda')['Precio'].transform(lambda x: x.rolling(6).mean())

colores = {'Tienda 1': 'blue', 'Tienda 2': 'brown', 'Tienda 3': 'green', 'Tienda 4': 'gold'}

plt.figure(figsize=(12, 8))
sns.lineplot(data=df_mensual, x='Fecha', y='Media_6M', hue='Tienda', palette=colores)

plt.title('Tendencia de ventas (media móvil de 6 meses)')
plt.ylabel('Ingresos suavizados (USD)')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
def calificacion(tienda):
  calificacion = tienda['Calificación'].mean()
  return calificacion
```

```python
calificaciones = {}

for i in range(len(tiendas)):
  calificaciones[tiendas[i]] = float(calificacion(datos[i]).round(2))

df_calificaciones = pd.DataFrame.from_dict(calificaciones, orient='index', columns=['Calificacion promedio'])
df_calificaciones
```

```python
df_calificaciones.plot(kind='line', color='green', legend=False, title='Calificación promedio')
plt.show()
```

# 4. Productos más y menos vendidos

```python
def mas_vendidos(tienda):
  mayores = tienda['Producto'].value_counts().head(3)
  return mayores

def menos_vendidos(tienda):
  menores = tienda['Producto'].value_counts().tail(3)
  return menores
```

```python
mayor_mayores = 0

for i in range(len(tiendas)):
  mayor_tienda = mas_vendidos(datos[i]).max()
  if mayor_tienda > mayor_mayores:
    mayor_mayores = mayor_tienda

fig, axes = plt.subplots(2,2, figsize=(12,8))
axes = axes.flatten()
colores = ['blue', 'brown', 'green', 'gold']

for i in range(len(tiendas)):
  top_3 = mas_vendidos(datos[i])
  display(top_3.to_frame(name=f'Más vendidos {tiendas[i]}'))
  graf = top_3.plot(kind='barh', color=colores[i], ylabel='', title=f'Tienda {i+1}', ax=axes[i])
  graf.invert_yaxis()
  graf.set_xlim(0, mayor_mayores * 1.1)

plt.suptitle('Productos más vendidos')
plt.tight_layout()
plt.show()
```

```python
mayor_menores = 0

for i in range(len(tiendas)):
  mayor_tienda = menos_vendidos(datos[i]).max()
  if mayor_tienda > mayor_menores:
    mayor_menores = mayor_tienda

fig, axes = plt.subplots(2,2, figsize=(12,8))
axes = axes.flatten()
colores = ['blue', 'brown', 'green', 'gold']

for i in range(len(tiendas)):
  bottom_3 = menos_vendidos(datos[i])
  display(bottom_3.to_frame(name=f'Menos vendidos {tiendas[i]}'))
  graf = bottom_3.plot(kind='barh', color=colores[i], ylabel='', title=f'Tienda {i+1}', ax=axes[i])
  graf.invert_yaxis()
  graf.set_xlim(0, mayor_menores * 1.1)

plt.suptitle('Productos menos vendidos')
plt.tight_layout()
plt.show()
```

# 6. Desempeño geográfico

```python
fig, axes = plt.subplots(2,2, figsize=(12,8))
axes = axes.flatten()
precio_mayor = max([d['Precio'].max() for d in datos])

for i in range(len(tiendas)):
  puntos = axes[i].scatter(
      datos[i]['lon'],
      datos[i]['lat'],
      c=datos[i]['Precio'],
      cmap='YlOrRd',
      alpha=0.6,
      vmin=0,
      vmax=precio_mayor
      )

  axes[i].set_title(f'Tienda {i+1}')
  axes[i].set_xlabel('Longitud')
  axes[i].set_ylabel('Latitud')
  plt.colorbar(puntos, ax=axes[i], label='Precio')

plt.suptitle('Desempeño geográfico')
plt.tight_layout()
plt.show()
```
